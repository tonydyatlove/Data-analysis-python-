```python
have_kids = []
for i in range(len(bookings.total_kids)):
    if bookings.total_kids[i] > 0:
        have_kids.append('True')
    else:
        have_kids.append('False')
```

```python
have_kids
```

```python
bookings['has_kids'] = have_kids
```
